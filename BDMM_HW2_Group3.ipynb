### Question 1:

The most typical use case for the database is to show information relating to a property listing to a customer. This is done by a query to the database which returns one of the listing documents. Currently a lot of time is spent when a listing is retrieved from the database to show to a customer. Decide what information should be returned in a typical query and optimise the structure for this use case. For example, we typically only want a sample of reviews but not all reviews (although all reviews ), the customer does not need to know past transaction data, etc. Update the document schema for this typical use case. This may involve the creation of new collections and documents.

* `TRANSFORMATION APPLIED:`
    * `Divide listingsAndReviews_new collection into 4 new smaller collections.`
* `REASONING BEHIND:`
    * `This division into more specific collections can help improve performance and simplify querying.`
* `EXPECTED RESULT:`
    * `HOSTS collection - With all the information about the hosts;`
    * `LISTING_REVIEWS collection - This is an intermediate collection with all the reviews per the listing so that then we can create a new collection called REVIEWS with each document being a single review;`
    * `TRANSACTIONS collection - With all the information about the transactions made per the listing;`
	* `LISTING collection - With all the information about the listing.`

```python
for document in db.listingsAndReviews_new.find():
    
    # Fields starting with "host"
    host_fields = {field: value for field, value in document.items() if field.startswith('host')}
    if host_fields:
        db.HOSTS.insert_one(host_fields)
        
    # Fields starting with "review"
    review_fields = {field: value for field, value in document.items() if field.startswith('reviews')}
    if review_fields:
        review_fields['_id'] = document['_id']
        db.LISTING_REVIEWS.insert_one(review_fields)
    
    # Fields starting with "transactions"
    transactions_fields = {field: value for field, value in document.items() if field.startswith('transactions')}
    if transactions_fields:
        transactions_fields['_id'] = document['_id']
        db.TRANSACTIONS.insert_one(transactions_fields)
        
    # All the other fields
    listing_fields = {field: value for field, value in document.items() if not field.startswith('host') 
                      and not field.startswith('reviews') and not field.startswith('transactions')}
    if listing_fields:
        listing_fields['host_id'] = document['host_id']
        db.LISTING.insert_one(listing_fields)
```

```python
def check_duplicates(collection):
    """
    This function checks for duplicate documents in a MongoDB collection by comparing all the fields, ignoring the 
    "_id", and deleting any duplicate found.
    
    input: Recieves a MongoDB collection.
    
    output: Information about if the colection has duplicates or not.
    """
    
    documents = list(collection.find())
    duplicates = 0

    # Loop over the documents
    for i, document1 in enumerate(documents):

        # Loop over the rest of the documents
        for j, document2 in enumerate(documents[i+1:]):

            # Ignoring the "_id" field
            document1_no_id = {field: value for field, value in document1.items() if field != '_id'}
            document2_no_id = {field: value for field, value in document2.items() if field != '_id'}

            # Check if there are duplicates
            if document1_no_id == document2_no_id:

                # Delete the duplicate document
                collection.delete_one({'_id': document2['_id']})

                duplicates += 1

    if duplicates != 0:           
        print("Two or more documents were found with exactly the same information.")
    else:
        print("No documents were found with exactly the same information.")
```

```python
pipeline = [{"$unwind": "$transactions"},
            {"$group": {"_id": "$_id",
                        "transactions": {"$push": "$transactions"},
                        "bucket_start_date": {"$first": "$bucket_start_date"},
                        "bucket_end_date": {"$first": "$bucket_end_date"}}},
            {"$addFields": {"transaction_count": {"$size": "$transactions"}}},
            {"$addFields": {"bucket_start_date": {"$min": "$transactions.date"}}},
            {"$addFields": {"bucket_end_date": {"$max": "$transactions.date"}}},
            {"$project": {"_id": 1,
                          "transactions": 1,
                          "transaction_count": 1,
                          "bucket_start_date": 1,
                          "bucket_end_date": 1}},
            {"$out": "TRANSACTIONS"}
           ]

db.TRANSACTIONS.aggregate(pipeline)
```

### Question 3:
Once per month we like to reward hosts with recognition. Pick three superhosts who have at least two property listings that can accommodate more than four people?

```python
pipeline = [{"$match": {"host_is_superhost": True, "host_listings_count": {"$gte": 2}}},
            {"$lookup": {"from": "LISTING", 
                         "localField": "host_id", 
                         "foreignField": "host_id", 
                         "as": "listing"}},
            {"$unwind": "$listing"},
            {"$match": {"listing.accommodates": {"$gt": 4}}},
            {"$sort": {"listing.review_scores.rating": -1}},
            {"$project": {"_id": 0,
                          "Host Name": "$host_name",
                          "Superhost": "$host_is_superhost",
                          "Properties": "$host_listings_count",
                          "Accommodates": "$listing.accommodates", 
                          "Reviews Rating": "$listing.review_scores.rating"}},
            {"$limit": 3}
           ]

result = db.HOSTS.aggregate(pipeline)

for document in result:
    pprint(document)
```

### Question 4:

```python
# Check if we can use market as city
unique_mismatches = set()

for i in db.LISTING.find({}):
    street_words = i['address']['street'].split(", ")
    if i["address"]["market"] != street_words[1]:
        mismatch = (i["address"]["market"], street_words[1])
        if mismatch not in unique_mismatches:
            unique_mismatches.add(mismatch)
            print(*mismatch, "MARKET:", i["address"]["market"])
```

```python
db.LISTING.create_index("address.market") # Indexing address market to improve performance in Q4 and Q5

pipeline = [{"$match": {"address.market": "New York", 
                        "amenities": {"$all": ["Waterfront", "Dishwasher"]}}},
            {"$group": {"_id": "$bed_type", "count": {"$sum": 1}}},
            {"$sort": {"count": -1}},
            {"$limit": 1}
           ]

result = list(db.LISTING.aggregate(pipeline))

print("The bed type most common in the listings that have waterfront and a dishwasher in New York is", 
      result[0]["_id"], end = ".")
```

For each property we store review scores across different metrics (accuracy, check-in, cleanliness etc). We are thinking we may soon add more metrics, although we don’t know what these will be.
```python
pipeline = [{"$project": {"listing_id": "$_id",
                          "review_scores": 1}},
            {"$addFields": {"review_scores_list": {"$objectToArray": "$review_scores"}}},
            {"$addFields": {"review_scores_sum": {"$sum": "$review_scores_list.v"}}},
            {"$addFields": {"review_scores_len": {"$size": "$review_scores_list"}}},
            {"$addFields": {"average_score": {"$cond": [{"$eq": ["$review_scores_len", 0]},
                                                        0,
                                                        {"$divide": ["$review_scores_sum", "$review_scores_len"]}]}}},
            {"$project": {"_id": 0,
                          "listing_id": 1,
                          "average_score": {"$round": ["$average_score", 2]}}},
            {"$limit": 5}
           ]

result = db.LISTING.aggregate(pipeline)

for document in result:
    print(f"Listing id: {document['listing_id']}, average overall review score: {document['average_score']}")
```

### Question 11:
We wish to be able to have better access to information about transaction, we wish to develop a search engine that can calculate the average value of transactions in a given period of time quickly for a given property.

```python
def transactions_function(listing_id, start_date, end_date):
    """
     This function calculates the average price of transactions of a given property in a certain period.
    
    Input:
    - listing_id - Property ID;
    - start_date - start date of the period;
    - end_date - start date of the period.
    
    Output: Average Price of transactions.
    """
    start = datetime.strptime(start_date, "%Y-%m-%d")
    end = datetime.strptime(end_date, "%Y-%m-%d")
    
    pipeline = [{"$match": {"_id": listing_id}},
                {"$unwind": "$transactions"},
                {"$match": {"transactions.date": {"$gte": start, "$lte": end}}},
                {"$group": {"_id": "$_id",
                            "avg_price": {"$avg": "$transactions.price"}}},
                {"$project": {"_id": 0,
                              "avg_price": {"$round": ["$avg_price", 2]}}}
               ]

    result = list(db.TRANSACTIONS.aggregate(pipeline))
    
    print(f"The listing with ID {listing_id} has an average value of transactions of {result[0]['avg_price']} from", 
          f"{start_date} to {end_date}.")
```
